```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.image as mpimg
import numpy as np
```

```python
x,y,z = np.loadtxt('output_001.dat',unpack=True)
N = int(len(z)**0.5)
z = z.reshape(N,N)
plt.imshow(z,extent=(np.amin(x),np.amax(x),np.amin(y),np.amax(y)),cmap='jet',interpolation='quadric')
plt.clim(0.26,1.58)
plt.title('T = 15.')
plt.xlabel('$q_x(r.l.u)$')
plt.ylabel('$q_y(r.l.u)$')
plt.colorbar()
plt.show()
```
