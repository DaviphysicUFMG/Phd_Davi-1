```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.image as mpimg
import numpy as np
```

```python
conf00 = mpimg.imread('00/ang00.png')
conf45 = mpimg.imread('45/ang45.png')
conf90 = mpimg.imread('90/ang90.png')
```

```python
x,y,a00 = np.loadtxt('00/ang00_T10.dat',unpack=True)
N1 = int(len(a00)**0.5)
a00 = a00.reshape(N1,N1)

_,_,a01 = np.loadtxt('00/ang00_T21.dat',unpack=True)
a01 = a01.reshape(N1,N1)

_,_,a02 = np.loadtxt('00/ang00_T07.dat',unpack=True)
a02 = a02.reshape(N1,N1)

_,_,a03 = np.loadtxt('00/ang00_T03.dat',unpack=True)
a03 = a03.reshape(N1,N1)

_,_,a04 = np.loadtxt('00/ang00_T01.dat',unpack=True)
a04 = a04.reshape(N1,N1)

##---------------------------------------------------

_,_,a10 = np.loadtxt('45/ang45_T10.dat',unpack=True)
a10 = a10.reshape(N1,N1)

_,_,a11 = np.loadtxt('45/ang45_T39.dat',unpack=True)
a11 = a11.reshape(N1,N1)

_,_,a12 = np.loadtxt('45/ang45_T11.dat',unpack=True)
a12 = a12.reshape(N1,N1)

_,_,a13 = np.loadtxt('45/ang45_T07.dat',unpack=True)
a13 = a13.reshape(N1,N1)

_,_,a14 = np.loadtxt('45/ang45_T01.dat',unpack=True)
a14 = a14.reshape(N1,N1)

##---------------------------------------------------

_,_,a20 = np.loadtxt('90/ang90_T10.dat',unpack=True)
a20 = a20.reshape(N1,N1)

_,_,a21 = np.loadtxt('90/ang90_T39.dat',unpack=True)
a21 = a21.reshape(N1,N1)

_,_,a22 = np.loadtxt('90/ang90_T23.dat',unpack=True)
a22 = a22.reshape(N1,N1)

_,_,a23 = np.loadtxt('90/ang90_T19.dat',unpack=True)
a23 = a23.reshape(N1,N1)

_,_,a24 = np.loadtxt('90/ang90_T11.dat',unpack=True)
a24 = a24.reshape(N1,N1)

##---------------------------------------------------

clim1 = (0.3,2)
clim2 = (0,10)
cmap1 = 'jet'
intpol = 'bicubic'#'quadric'
tamanho = (np.amin(x),np.amax(x),np.amin(y),np.amax(y))
```

```python
##
fig,axes = plt.subplots(3,6,dpi=200,figsize=(20, 10))
axes[0,0].imshow(conf00)
axes[1,0].imshow(conf45)
axes[2,0].imshow(conf90)
axes[0,0].axis('off')
axes[1,0].axis('off')
axes[2,0].axis('off')
##
##
axes[0,1].imshow(a00,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,0.85))
axes[0,2].imshow(a01,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,1.4))
axes[0,3].imshow(a02,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,1.5))
axes[0,4].imshow(a03,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,8))
axes[0,5].imshow(a04,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,100))
##
axes[1,1].imshow(a10,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,0.65))
axes[1,2].imshow(a11,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,0.75))
axes[1,3].imshow(a12,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,1.5))
axes[1,4].imshow(a13,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,1.5))
axes[1,5].imshow(a14,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,1.5))
##
axes[2,1].imshow(a20,extent=tamanho,cmap=cmap1,interpolation=intpol)#,clim=(0,0.8))
axes[2,2].imshow(a21,extent=tamanho,cmap=cmap1,interpolation=intpol)#,clim=(.3,1.4))
axes[2,3].imshow(a22,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,2))
axes[2,4].imshow(a23,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,2))
axes[2,5].imshow(a24,extent=tamanho,cmap=cmap1,interpolation=intpol,clim=(0,5))

plt.savefig('mssf.pdf')
```
